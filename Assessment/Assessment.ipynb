```python
#Creating a new dataframe from the existing dataframe

data_table = data.copy()

#Renaming dataframe columns

data_table.rename(columns = {'Customer_Name':'Name', 'Customer_Id':'Cust_I', 'Open_Date':'Open_Dt', 
                             'Last_Consulted_Date':'Consul_Dt', 'Vaccination_Id':'VAC_ID', 'Is_Active':'FLAG'})
```

```python
# Removing extra spaces from the string values

#data_table['Name'] = data_table['Name'].str.strip()
#data_table['VAC_ID'] = data_table['VAC_ID'].str.strip()
data_table['Dr_Name'] = data_table['Dr_Name'].str.strip()
data_table['State'] = data_table['State'].str.strip()
data_table['Country'] = data_table['Country'].str.strip()
```

```python
# Country Specific data

TABLE_INDIA = data_table[data_table['Country']=='IND']
TABLE_USA = data_table[data_table['Country']=='USA']
TABLE_PHILIPPINES = data_table[data_table['Country']=='PHIL']
TABLE_NEWYORK = data_table[data_table['Country']=='NYC']
TABLE_AUSTRALIA = data_table[data_table['Country']=='AU']
```
